```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pairs import get_pairs
import pickle
```

## Relações Anteriores, sem shuffling 

```python
pairs_old = pickle.load(open("influence_pairs.pkl", "rb"))
```

```python
ax = sns.heatmap(pairs_old, cmap="RdYlGn")
plt.title("Dependência de questões ENEM 2015 CN Azul \n (y depende de x?)")
plt.xlabel("Pré Requisito")
plt.ylabel("Questão")
plt.show()
```

## Relações com shuffling

```python
pairs = get_pairs(45)
```

```python
ax = sns.heatmap(pairs, cmap="RdYlGn")
plt.title("Dependência de questões ENEM 2015 CN Azul \n (y depende de x?)")
plt.xlabel("Pré Requisito")
plt.ylabel("Questão")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
c="RdYlGn"
sns.heatmap(pairs_old, cmap=c, ax=ax1)
ax1.set_title("Dependência de questões ENEM 2015 CN Azul \n Sem shuffling")
ax1.set_xlabel("Pré Requisito")
ax1.set_ylabel("Questão")

sns.heatmap(pairs, cmap=c, ax=ax2)
ax2.set_title("Dependência de questões ENEM 2015 CN Azul \n Com shuffling")
ax2.set_xlabel("Pré Requisito")
ax2.set_ylabel("Questão")
plt.tight_layout()
# plt.savefig("pairs_shuffling.jpg", dpi=400)
plt.show()
```
